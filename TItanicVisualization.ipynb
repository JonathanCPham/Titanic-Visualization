## The Data
Today I will be using seaborn and the titanic dataset to show some visualizations.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
df = sns.load_dataset('titanic')
df.shape
```

```python
df.info()
```

```python
df.head()
```

We want to check the missing values

```python
df.isnull().sum()
```

```python
# Heatmap that displays all missing data in the titanic dataset.
sns.heatmap(df.isnull(), yticklabels=False, cbar=False)
```

```python
# Showing the distribution of fare prices and age of the passengers.
sns.histplot(data=df, x='fare', bins=30)
plt.ylabel('Frequency')
plt.title('fare distribution with hist')
plt.show()
plt.ylabel('Frequency')
plt.title('age distribution with hist')
sns.histplot(data=df, x='age', bins=30)
plt.show()
```

```python
# Countplot counts the each category of value and plot that.
sns.countplot(data=df, x='survived')
plt.show()
```

```python
#Bar graph showing different passenger classes.
sns.countplot(data=df, x='pclass')
plt.show()
```

```python
# Joint plot showing distribution of fare prices in comparison to age
sns.jointplot(data=df, x='fare', y='age', kind='scatter')
plt.xlim(right=600)
plt.ylim(top=90)
```

```python
#Showing distribution of tickets bought at x amount of fare price
plt.hist(df['fare'], bins=25)
plt.xlim(left=0)
plt.xlim(right=600)
plt.ylim(top=500)
plt.xlabel('fare')
```

```python
#Boxplot displaying average age in comparison to passenger class
sns.boxplot(data=df, x='class', y='age')
```

```python
#A swarmplot showing age range in comparison to passenger class.
sns.swarmplot(data=df, x='class',y='age')
```

```python
#A bar graph showing passenger genders.
sns.countplot(data=df, x='sex')
plt.show()
```

```python
#A titanic heatmap showing probablily of surival.
titanic = sns.load_dataset('titanic')
data = titanic[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']]
corr = data.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
#A heatmap of passengers in different age ranges for male and female.
age = sns.FacetGrid(titanic, col="sex", height=5)
age.map(plt.hist, 'age')
```
